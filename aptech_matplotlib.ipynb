```python
fig=py.figure(figsize=(8,6))
py.plot(x,y,color='r',linewidth=3,linestyle='--',marker='o',markersize=15,markerfacecolor='green',
       markeredgewidth=4,markeredgecolor='yellow')
```

```python


fig=py.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='r',linewidth=3,linestyle='-', marker='o', markersize=20, markerfacecolor='yellow'
       ,markeredgewidth=3,markeredgecolor='green')
#color='r','b','o','black','g','y'
#linewidth can take any value the higher the value the thicker ad width
#linestyle can be '--', dashed, ':', ''
#marker can be 'o','1','+'
```

```python
fig,ax=py.subplots(figsize=(12,6))
ax.plot(x,x+1,color='red',linewidth=0.25)
ax.plot(x,x+2,color='red',linewidth=0.50)
ax.plot(x,x+3,color='red',linewidth=0.1)
ax.plot(x,x+4,color='red',linewidth=0.2)



ax.plot(x,x+5,color='green',linewidth=3,linestyle='-')
ax.plot(x,x+6,color='green',linewidth=3,linestyle='-.')
ax.plot(x,x+7,color='green',linewidth=3,linestyle=':')


ax.plot(x,x+9,color='blue',linewidth=3,ls='-',marker='+')
ax.plot(x,x+10,color='blue',linewidth=3,ls='--',marker='o')
ax.plot(x,x+11,color='blue',linewidth=3,ls='-',marker='s')
ax.plot(x,x+12,color='blue',linewidth=3,ls='--',marker='1')
```
